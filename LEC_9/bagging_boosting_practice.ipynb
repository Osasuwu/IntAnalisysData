## Часть 1: Подготовка данных и импорт библиотек

Начнем с импорта необходимых библиотек и загрузки данных для экспериментов.

```python
# Импорт основных библиотек
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_wine, load_breast_cancer, make_classification
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV, validation_curve
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, roc_auc_score
import warnings
warnings.filterwarnings('ignore')

# Настройка визуализации
plt.style.use('seaborn-v0_8')
plt.rcParams['figure.figsize'] = (12, 8)
sns.set_palette("husl")

print("✅ Библиотеки успешно импортированы!")
print(f"Версия NumPy: {np.__version__}")
print(f"Версия Pandas: {pd.__version__}")
```

```python
# Импорт ансамблевых методов
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import (
    BaggingClassifier,
    RandomForestClassifier,
    ExtraTreesClassifier,
    AdaBoostClassifier,
    GradientBoostingClassifier
)

# Попробуем импортировать XGBoost (если установлен)
try:
    from xgboost import XGBClassifier
    XGBOOST_AVAILABLE = True
    print("✅ XGBoost доступен")
except ImportError:
    XGBOOST_AVAILABLE = False
    print("⚠️ XGBoost не установлен. Установите командой: pip install xgboost")

print("✅ Ансамблевые методы импортированы!")
```

### Задание 1.1: Загрузка и исследование данных

Загрузите датасет Wine и изучите его структуру.

```python
# Загрузка датасета Wine для классификации сортов вина
wine_data = load_wine()
X_wine, y_wine = wine_data.data, wine_data.target

# Создание DataFrame для удобства анализа и визуализации
wine_df = pd.DataFrame(X_wine, columns=wine_data.feature_names)
wine_df['target'] = y_wine
wine_df['target_name'] = [wine_data.target_names[i] for i in y_wine]

# Вывод базовой статистики датасета
print("Информация о датасете Wine:")
print(f"Размер: {wine_df.shape}")
print(f"Признаки: {len(wine_data.feature_names)}")
print(f"Классы: {wine_data.target_names}")
print(f"Распределение классов:\n{wine_df['target_name'].value_counts()}")

# Покажите первые строки
wine_df.head()
```

```python
# Визуализация распределения классов
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

# Гистограмма распределения классов для оценки баланса данных
wine_df['target_name'].value_counts().plot(kind='bar', ax=axes[0], color='skyblue')
axes[0].set_title('Распределение классов в датасете Wine')
axes[0].set_xlabel('Класс вина')
axes[0].set_ylabel('Количество образцов')
axes[0].tick_params(axis='x', rotation=45)

# Круговая диаграмма для визуализации пропорций классов
wine_df['target_name'].value_counts().plot(kind='pie', ax=axes[1], autopct='%1.1f%%')
axes[1].set_title('Процентное распределение классов')
axes[1].set_ylabel('')

plt.tight_layout()
plt.show()

# Проверка качества данных: наличие пропущенных значений
print(f"\nПропущенные значения: {wine_df.isnull().sum().sum()}")
```

### Задание 1.2: Подготовка данных для обучения

Разделите данные на обучающую и тестовую выборки.

```python
# Разделение данных на признаки и целевую переменную
X = wine_df.drop(['target', 'target_name'], axis=1)
y = wine_df['target']

# Разбиение на обучающую (70%) и тестовую (30%) выборки с сохранением пропорций классов
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

print(f"Размер обучающей выборки: {X_train.shape}")
print(f"Размер тестовой выборки: {X_test.shape}")

# Стандартизация признаков для улучшения работы моделей
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print("✅ Данные подготовлены и стандартизированы!")
```

## Часть 2: Реализация Bagging методов

Теперь реализуем и сравним различные Bagging методы.

### Задание 2.1: Сравнение одиночного дерева с Bagging

Сравните производительность одиночного дерева решений с Bagging ансамблем.

```python
# Создание базовой модели - одиночное дерево решений
single_tree = DecisionTreeClassifier(random_state=42)

# Создание Bagging ансамбля из 100 деревьев для снижения дисперсии
bagging_clf = BaggingClassifier(
    estimator=DecisionTreeClassifier(),
    n_estimators=100,
    random_state=42
)

# Обучение моделей на стандартизированных данных
single_tree.fit(X_train_scaled, y_train)
bagging_clf.fit(X_train_scaled, y_train)

# Получение предсказаний на тестовой выборке
single_pred = single_tree.predict(X_test_scaled)
bagging_pred = bagging_clf.predict(X_test_scaled)

# Вычисление метрик точности для сравнения моделей
single_accuracy = accuracy_score(y_test, single_pred)
bagging_accuracy = accuracy_score(y_test, bagging_pred)

print("Сравнение производительности:")
print(f"Одиночное дерево решений: {single_accuracy:.4f}")
print(f"Bagging (100 деревьев): {bagging_accuracy:.4f}")
print(f"Улучшение: {bagging_accuracy - single_accuracy:.4f}")
```

### Задание 2.2: Random Forest и Extra Trees

Реализуйте Random Forest и Extra Trees, сравните их производительность.

```python
# Создание Random Forest с 100 деревьями и случайным выбором признаков
rf_clf = RandomForestClassifier(
    n_estimators=100,
    random_state=42
)

# Создание Extra Trees с полностью случайными разбиениями в узлах
et_clf = ExtraTreesClassifier(
    n_estimators=100,
    random_state=42
)

# Оценка моделей с помощью 5-fold кросс-валидации
rf_scores = cross_val_score(rf_clf, X_train_scaled, y_train, cv=5, scoring='accuracy')
et_scores = cross_val_score(et_clf, X_train_scaled, y_train, cv=5, scoring='accuracy')

print("Результаты кросс-валидации:")
print(f"Random Forest: {rf_scores.mean():.4f} ± {rf_scores.std():.4f}")
print(f"Extra Trees: {et_scores.mean():.4f} ± {et_scores.std():.4f}")

# Обучение финальных моделей на полной обучающей выборке
rf_clf.fit(X_train_scaled, y_train)
et_clf.fit(X_train_scaled, y_train)

# Генерация предсказаний для оценки на тестовых данных
rf_pred = rf_clf.predict(X_test_scaled)
et_pred = et_clf.predict(X_test_scaled)

print("\nТочность на тестовой выборке:")
print(f"Random Forest: {accuracy_score(y_test, rf_pred):.4f}")
print(f"Extra Trees: {accuracy_score(y_test, et_pred):.4f}")
```

### Задание 2.3: Анализ важности признаков в Random Forest

Проанализируйте важность признаков и визуализируйте результаты.

```python
# Извлечение важности признаков из обученной модели Random Forest
feature_importance = pd.DataFrame({
    'feature': X.columns,
    'importance': rf_clf.feature_importances_
}).sort_values('importance', ascending=False)

# Визуализация топ-10 наиболее влиятельных признаков
plt.figure(figsize=(12, 8))
sns.barplot(data=feature_importance.head(10), y='feature', x='importance', palette='viridis')
plt.title('Топ 10 наиболее важных признаков (Random Forest)', fontsize=16, fontweight='bold')
plt.xlabel('Важность признака', fontsize=12)
plt.ylabel('Признак', fontsize=12)
plt.tight_layout()
plt.show()

# Вывод численных значений важности для топ-5 признаков
print("Топ 5 наиболее важных признаков:")
for i, row in feature_importance.head().iterrows():
    print(f"{row['feature']}: {row['importance']:.4f}")
```

## Часть 3: Реализация Boosting методов

Теперь реализуем различные Boosting алгоритмы.

### Задание 3.1: AdaBoost

Реализуйте AdaBoost и исследуйте влияние количества деревьев на производительность.

```python
# Исследование зависимости производительности AdaBoost от числа базовых оценок
n_estimators_range = [1, 5, 10, 25, 50, 100, 200]
ada_train_scores = []
ada_test_scores = []

for n_est in n_estimators_range:
    # Создание AdaBoost классификатора с пнями решений (деревья глубины 1)
    ada_clf = AdaBoostClassifier(
        estimator=DecisionTreeClassifier(max_depth=1),
        n_estimators=n_est,
        random_state=42,
        algorithm='SAMME'
    )
    
    # Обучение модели на текущем количестве итераций
    ada_clf.fit(X_train_scaled, y_train)
    
    # Оценка точности на обучающей и тестовой выборках
    train_pred = ada_clf.predict(X_train_scaled)
    test_pred = ada_clf.predict(X_test_scaled)
    
    ada_train_scores.append(accuracy_score(y_train, train_pred))
    ada_test_scores.append(accuracy_score(y_test, test_pred))

# Построение кривых обучения для анализа переобучения
plt.figure(figsize=(10, 6))
plt.plot(n_estimators_range, ada_train_scores, 'o-', label='Обучающая выборка', linewidth=2)
plt.plot(n_estimators_range, ada_test_scores, 'o-', label='Тестовая выборка', linewidth=2)
plt.xlabel('Количество базовых оценок')
plt.ylabel('Точность')
plt.title('Кривые обучения AdaBoost')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

# Найдите оптимальное количество деревьев
optimal_n = n_estimators_range[np.argmax(ada_test_scores)]
best_score = max(ada_test_scores)
print(f"Оптимальное количество деревьев: {optimal_n}")
print(f"Лучшая точность: {best_score:.4f}")
```

### Задание 3.2: Gradient Boosting

Реализуйте Gradient Boosting и исследуйте влияние learning rate.

```python
# Исследование влияния скорости обучения на производительность Gradient Boosting
learning_rates = [0.01, 0.05, 0.1, 0.2, 0.5, 1.0]
gb_results = []

for lr in learning_rates:
    # Создание Gradient Boosting с различными значениями learning_rate
    gb_clf = GradientBoostingClassifier(
        n_estimators=100,
        learning_rate=lr,
        max_depth=3,
        random_state=42
    )
    
    # Оценка производительности через кросс-валидацию
    scores = cross_val_score(gb_clf, X_train_scaled, y_train, cv=5, scoring='accuracy')
    
    gb_results.append({
        'learning_rate': lr,
        'mean_score': scores.mean(),
        'std_score': scores.std()
    })

# Визуализация зависимости точности от скорости обучения
gb_df = pd.DataFrame(gb_results)
plt.figure(figsize=(10, 6))
plt.errorbar(gb_df['learning_rate'], gb_df['mean_score'], 
             yerr=gb_df['std_score'], fmt='bo-', linewidth=2, markersize=8, capsize=5)
plt.xlabel('Learning Rate')
plt.ylabel('Точность (Cross-validation)')
plt.title('Влияние Learning Rate на производительность Gradient Boosting')
plt.grid(True, alpha=0.3)
plt.xscale('log')
plt.show()

print("Результаты для различных Learning Rate:")
for result in gb_results:
    print(f"LR = {result['learning_rate']}: {result['mean_score']:.4f} ± {result['std_score']:.4f}")
```

### Задание 3.3: XGBoost (если доступен)

Если XGBoost установлен, протестируйте его производительность.

```python
if XGBOOST_AVAILABLE:
    # Создание XGBoost классификатора с оптимизированными параметрами
    xgb_clf = XGBClassifier(
        n_estimators=100,
        learning_rate=0.1,
        max_depth=3,
        random_state=42,
        eval_metric='mlogloss'
    )
    
    # Кросс-валидационная оценка производительности XGBoost
    xgb_scores = cross_val_score(xgb_clf, X_train_scaled, y_train, cv=5, scoring='accuracy')
    print(f"XGBoost CV Score: {xgb_scores.mean():.4f} ± {xgb_scores.std():.4f}")
    
    # Обучение финальной модели XGBoost на полных данных
    xgb_clf.fit(X_train_scaled, y_train)
    xgb_pred = xgb_clf.predict(X_test_scaled)
    xgb_accuracy = accuracy_score(y_test, xgb_pred)
    print(f"XGBoost Test Accuracy: {xgb_accuracy:.4f}")
else:
    print("XGBoost не доступен. Установите командой: pip install xgboost")
```

## Часть 4: Комплексное сравнение методов

Сравним все реализованные методы на одном датасете.

### Задание 4.1: Сравнение всех методов

Создайте сравнительную таблицу производительности всех методов.

```python
# Создание словаря со всеми ансамблевыми моделями для комплексного сравнения
models = {
    'Single Decision Tree': DecisionTreeClassifier(random_state=42),
    'Bagging': BaggingClassifier(estimator=DecisionTreeClassifier(), 
                               n_estimators=100, random_state=42),
    'Random Forest': RandomForestClassifier(n_estimators=100, random_state=42),
    'Extra Trees': ExtraTreesClassifier(n_estimators=100, random_state=42),
    'AdaBoost': AdaBoostClassifier(estimator=DecisionTreeClassifier(max_depth=1),
                                 n_estimators=100, random_state=42, algorithm='SAMME'),
    'Gradient Boosting': GradientBoostingClassifier(n_estimators=100, learning_rate=0.1,
                                                   max_depth=3, random_state=42)
}

# Добавим XGBoost если доступен
if XGBOOST_AVAILABLE:
    models['XGBoost'] = XGBClassifier(n_estimators=100, learning_rate=0.1, max_depth=3,
                                     random_state=42, eval_metric='mlogloss')

# Комплексная оценка всех моделей с кросс-валидацией и тестированием
results = []
for name, model in models.items():
    print(f"Обучение {name}...")
    
    # 5-fold кросс-валидация для оценки обобщающей способности
    cv_scores = cross_val_score(model, X_train_scaled, y_train, cv=5, scoring='accuracy')
    
    # Обучение модели на полной обучающей выборке
    model.fit(X_train_scaled, y_train)
    test_pred = model.predict(X_test_scaled)
    test_accuracy = accuracy_score(y_test, test_pred)
    
    # Вычисление AUC-ROC для многоклассовой классификации
    try:
        test_pred_proba = model.predict_proba(X_test_scaled)
        auc_score = roc_auc_score(y_test, test_pred_proba, multi_class='ovr')
    except:
        auc_score = np.nan
    
    results.append({
        'Model': name,
        'CV Mean': cv_scores.mean(),
        'CV Std': cv_scores.std(),
        'Test Accuracy': test_accuracy,
        'AUC Score': auc_score
    })

# Создание сводной таблицы результатов с сортировкой по точности
results_df = pd.DataFrame(results)
results_df = results_df.sort_values('Test Accuracy', ascending=False)

print("\n" + "="*80)
print("СРАВНЕНИЕ ПРОИЗВОДИТЕЛЬНОСТИ МОДЕЛЕЙ")
print("="*80)
print(results_df.to_string(index=False, float_format='%.4f'))
```

### Задание 4.2: Визуализация сравнения

Создайте визуализации для сравнения производительности моделей.

```python
# Создание комплексной визуализации для сравнения всех моделей
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# График 1: Сравнение точности кросс-валидации
ax1 = axes[0, 0]
bars1 = ax1.barh(results_df['Model'], results_df['CV Mean'], color='skyblue')
ax1.set_xlabel('CV Accuracy')
ax1.set_title('Точность кросс-валидации', fontweight='bold')
ax1.grid(True, alpha=0.3)
# Аннотация значений точности на графике
for i, bar in enumerate(bars1):
    width = bar.get_width()
    ax1.text(width + 0.001, bar.get_y() + bar.get_height()/2, 
             f'{width:.3f}', ha='left', va='center', fontsize=9)

# График 2: Точность на тестовой выборке
ax2 = axes[0, 1]
bars2 = ax2.barh(results_df['Model'], results_df['Test Accuracy'], color='lightcoral')
ax2.set_xlabel('Test Accuracy')
ax2.set_title('Точность на тестовой выборке', fontweight='bold')
ax2.grid(True, alpha=0.3)
# Аннотация значений на графике для улучшения читаемости
for i, bar in enumerate(bars2):
    width = bar.get_width()
    ax2.text(width + 0.001, bar.get_y() + bar.get_height()/2, 
             f'{width:.3f}', ha='left', va='center', fontsize=9)

# График 3: Оценка AUC-ROC для моделей с вероятностными предсказаниями
ax3 = axes[1, 0]
auc_data = results_df.dropna(subset=['AUC Score'])
if not auc_data.empty:
    bars3 = ax3.barh(auc_data['Model'], auc_data['AUC Score'], color='lightgreen')
    ax3.set_xlabel('AUC Score')
    ax3.set_title('AUC Score', fontweight='bold')
    ax3.grid(True, alpha=0.3)
    for i, bar in enumerate(bars3):
        width = bar.get_width()
        ax3.text(width + 0.001, bar.get_y() + bar.get_height()/2, 
                 f'{width:.3f}', ha='left', va='center', fontsize=9)
else:
    ax3.text(0.5, 0.5, 'AUC Score недоступен', ha='center', va='center', transform=ax3.transAxes)

# График 4: Сравнение CV и Test точности
ax4 = axes[1, 1]
x_pos = np.arange(len(results_df))
ax4.bar(x_pos - 0.2, results_df['CV Mean'], 0.4, label='CV Accuracy', alpha=0.7)
ax4.bar(x_pos + 0.2, results_df['Test Accuracy'], 0.4, label='Test Accuracy', alpha=0.7)
ax4.set_xlabel('Модели')
ax4.set_ylabel('Точность')
ax4.set_title('Сравнение CV и Test точности', fontweight='bold')
ax4.set_xticks(x_pos)
ax4.set_xticklabels(results_df['Model'], rotation=45, ha='right')
ax4.legend()
ax4.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

## Часть 5: Настройка гиперпараметров

Настроим гиперпараметры для лучшей модели.

### Задание 5.1: GridSearchCV для Random Forest

Используйте GridSearchCV для оптимизации гиперпараметров Random Forest.

```python
# Определение пространства поиска гиперпараметров для Random Forest
rf_param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [3, 5, 10, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['sqrt', 'log2', None]
}

print(f"Общее количество комбинаций: {len(rf_param_grid['n_estimators']) * len(rf_param_grid['max_depth']) * len(rf_param_grid['min_samples_split']) * len(rf_param_grid['min_samples_leaf']) * len(rf_param_grid['max_features'])}")

# Упростим сетку для быстрого выполнения
rf_param_grid_simple = {
    'n_estimators': [50, 100],
    'max_depth': [5, 10, None],
    'min_samples_split': [2, 5],
    'max_features': ['sqrt', 'log2']
}

print(f"Упрощенная сетка: {len(rf_param_grid_simple['n_estimators']) * len(rf_param_grid_simple['max_depth']) * len(rf_param_grid_simple['min_samples_split']) * len(rf_param_grid_simple['max_features'])} комбинаций")

# Создание GridSearchCV для автоматической оптимизации гиперпараметров
rf_grid_search = GridSearchCV(
    RandomForestClassifier(random_state=42),
    rf_param_grid_simple,
    cv=5,
    scoring='accuracy',
    n_jobs=-1,  # Параллельное выполнение на всех ядрах процессора
    verbose=1
)

print("Запуск Grid Search для Random Forest...")
# Выполнение исчерпывающего поиска по сетке параметров
rf_grid_search.fit(X_train_scaled, y_train)

print("\nЛучшие параметры Random Forest:")
print(rf_grid_search.best_params_)
print(f"Лучший CV score: {rf_grid_search.best_score_:.4f}")

# Оценка оптимизированной модели на отложенной тестовой выборке
best_rf = rf_grid_search.best_estimator_
rf_optimized_pred = best_rf.predict(X_test_scaled)
rf_optimized_accuracy = accuracy_score(y_test, rf_optimized_pred)

print(f"Точность оптимизированной модели на тесте: {rf_optimized_accuracy:.4f}")
```

### Задание 5.2: Кривые валидации

### Задание 6.2: Ваш анализ

Проведите полный анализ датасета Breast Cancer, включая:
1. Сравнение всех ансамблевых методов
2. Настройку гиперпараметров для лучшей модели
3. Анализ важности признаков
4. Выводы и рекомендации

```python
"""
=================================================================================
ПОЛНЫЙ АНАЛИЗ ДАТАСЕТА BREAST CANCER
=================================================================================
Выполняется комплексный анализ с применением всех изученных ансамблевых методов
"""

# ============================================================================
# 1. СРАВНЕНИЕ ВСЕХ АНСАМБЛЕВЫХ МЕТОДОВ
# ============================================================================
print("="*80)
print("1. СРАВНЕНИЕ АНСАМБЛЕВЫХ МЕТОДОВ НА ДАТАСЕТЕ BREAST CANCER")
print("="*80)

models_cancer = {
    'Single Decision Tree': DecisionTreeClassifier(random_state=42),
    'Bagging': BaggingClassifier(
        estimator=DecisionTreeClassifier(), 
        n_estimators=100, 
        random_state=42
    ),
    'Random Forest': RandomForestClassifier(
        n_estimators=100, 
        random_state=42
    ),
    'Extra Trees': ExtraTreesClassifier(
        n_estimators=100, 
        random_state=42
    ),
    'AdaBoost': AdaBoostClassifier(
        estimator=DecisionTreeClassifier(max_depth=1),
        n_estimators=100, 
        random_state=42,
        algorithm='SAMME'
    ),
    'Gradient Boosting': GradientBoostingClassifier(
        n_estimators=100, 
        learning_rate=0.1,
        max_depth=3, 
        random_state=42
    ),
    'XGBoost': XGBClassifier(
        n_estimators=100, 
        learning_rate=0.1, 
        max_depth=3,
        random_state=42, 
        eval_metric='logloss'
    )
}

# Оценка всех моделей
cancer_results = []
for name, model in models_cancer.items():
    print(f"Обучение и оценка {name}...")
    
    # Кросс-валидация на обучающей выборке
    cv_scores = cross_val_score(
        model, X_train_cancer_scaled, y_train_cancer, 
        cv=5, scoring='accuracy', n_jobs=-1
    )
    
    # Обучение на полной обучающей выборке
    model.fit(X_train_cancer_scaled, y_train_cancer)
    
    # Предсказания на тестовой выборке
    test_pred = model.predict(X_test_cancer_scaled)
    test_accuracy = accuracy_score(y_test_cancer, test_pred)
    
    # Вычисление AUC-ROC
    try:
        test_pred_proba = model.predict_proba(X_test_cancer_scaled)
        auc_score = roc_auc_score(y_test_cancer, test_pred_proba[:, 1])
    except:
        auc_score = np.nan
    
    cancer_results.append({
        'Model': name,
        'CV Mean': cv_scores.mean(),
        'CV Std': cv_scores.std(),
        'Test Accuracy': test_accuracy,
        'AUC-ROC': auc_score
    })

# Создание DataFrame с результатами
cancer_results_df = pd.DataFrame(cancer_results)
cancer_results_df = cancer_results_df.sort_values('Test Accuracy', ascending=False)

print("\n" + "="*80)
print("РЕЗУЛЬТАТЫ СРАВНЕНИЯ МОДЕЛЕЙ")
print("="*80)
print(cancer_results_df.to_string(index=False, float_format='%.4f'))

# ============================================================================
# 2. ВИЗУАЛИЗАЦИЯ РЕЗУЛЬТАТОВ
# ============================================================================
print("\n" + "="*80)
print("2. ВИЗУАЛИЗАЦИЯ ПРОИЗВОДИТЕЛЬНОСТИ МОДЕЛЕЙ")
print("="*80)

fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# График 1: CV Accuracy
ax1 = axes[0, 0]
colors_cv = plt.cm.viridis(np.linspace(0, 1, len(cancer_results_df)))
bars1 = ax1.barh(cancer_results_df['Model'], cancer_results_df['CV Mean'], color=colors_cv)
ax1.set_xlabel('CV Accuracy', fontsize=12, fontweight='bold')
ax1.set_title('Кросс-валидация (5-fold)', fontsize=14, fontweight='bold')
ax1.grid(True, alpha=0.3, axis='x')
for i, (bar, val) in enumerate(zip(bars1, cancer_results_df['CV Mean'])):
    ax1.text(val + 0.005, bar.get_y() + bar.get_height()/2, 
             f'{val:.4f}', ha='left', va='center', fontsize=9)

# График 2: Test Accuracy
ax2 = axes[0, 1]
colors_test = plt.cm.plasma(np.linspace(0, 1, len(cancer_results_df)))
bars2 = ax2.barh(cancer_results_df['Model'], cancer_results_df['Test Accuracy'], color=colors_test)
ax2.set_xlabel('Test Accuracy', fontsize=12, fontweight='bold')
ax2.set_title('Точность на тестовой выборке', fontsize=14, fontweight='bold')
ax2.grid(True, alpha=0.3, axis='x')
for i, (bar, val) in enumerate(zip(bars2, cancer_results_df['Test Accuracy'])):
    ax2.text(val + 0.005, bar.get_y() + bar.get_height()/2, 
             f'{val:.4f}', ha='left', va='center', fontsize=9)

# График 3: AUC-ROC Score
ax3 = axes[1, 0]
auc_data = cancer_results_df.dropna(subset=['AUC-ROC'])
colors_auc = plt.cm.coolwarm(np.linspace(0, 1, len(auc_data)))
bars3 = ax3.barh(auc_data['Model'], auc_data['AUC-ROC'], color=colors_auc)
ax3.set_xlabel('AUC-ROC Score', fontsize=12, fontweight='bold')
ax3.set_title('ROC AUC Score', fontsize=14, fontweight='bold')
ax3.grid(True, alpha=0.3, axis='x')
for i, (bar, val) in enumerate(zip(bars3, auc_data['AUC-ROC'])):
    ax3.text(val + 0.005, bar.get_y() + bar.get_height()/2, 
             f'{val:.4f}', ha='left', va='center', fontsize=9)

# График 4: Сравнение CV vs Test Accuracy
ax4 = axes[1, 1]
x_pos = np.arange(len(cancer_results_df))
width = 0.35
ax4.bar(x_pos - width/2, cancer_results_df['CV Mean'], width, 
        label='CV Accuracy', alpha=0.8, color='steelblue')
ax4.bar(x_pos + width/2, cancer_results_df['Test Accuracy'], width, 
        label='Test Accuracy', alpha=0.8, color='coral')
ax4.set_ylabel('Accuracy', fontsize=12, fontweight='bold')
ax4.set_title('CV vs Test Accuracy', fontsize=14, fontweight='bold')
ax4.set_xticks(x_pos)
ax4.set_xticklabels(cancer_results_df['Model'], rotation=45, ha='right')
ax4.legend(fontsize=10)
ax4.grid(True, alpha=0.3, axis='y')

plt.tight_layout()
plt.show()

# ============================================================================
# 3. НАСТРОЙКА ГИПЕРПАРАМЕТРОВ ДЛЯ ЛУЧШЕЙ МОДЕЛИ
# ============================================================================
print("\n" + "="*80)
print("3. ОПТИМИЗАЦИЯ ГИПЕРПАРАМЕТРОВ (RANDOM FOREST)")
print("="*80)

# Определение сетки параметров
param_grid_cancer = {
    'n_estimators': [50, 100, 150],
    'max_depth': [5, 10, 15, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

print(f"Всего комбинаций: {np.prod([len(v) for v in param_grid_cancer.values()])}")
print("Запуск GridSearchCV...")

# GridSearchCV
grid_search_cancer = GridSearchCV(
    RandomForestClassifier(random_state=42),
    param_grid_cancer,
    cv=5,
    scoring='accuracy',
    n_jobs=-1,
    verbose=1
)

grid_search_cancer.fit(X_train_cancer_scaled, y_train_cancer)

print(f"\nЛучшие параметры: {grid_search_cancer.best_params_}")
print(f"Лучший CV score: {grid_search_cancer.best_score_:.4f}")

# Оценка оптимизированной модели
best_model_cancer = grid_search_cancer.best_estimator_
test_pred_optimized = best_model_cancer.predict(X_test_cancer_scaled)
test_accuracy_optimized = accuracy_score(y_test_cancer, test_pred_optimized)
print(f"Точность оптимизированной модели на тесте: {test_accuracy_optimized:.4f}")

# Отчет о классификации
print("\nОтчет о классификации (оптимизированная модель):")
print(classification_report(y_test_cancer, test_pred_optimized, 
                          target_names=cancer_data.target_names))

# ============================================================================
# 4. АНАЛИЗ ВАЖНОСТИ ПРИЗНАКОВ
# ============================================================================
print("\n" + "="*80)
print("4. АНАЛИЗ ВАЖНОСТИ ПРИЗНАКОВ")
print("="*80)

# Получение важности признаков
feature_importance_cancer = pd.DataFrame({
    'feature': X_cancer_clean.columns,
    'importance': best_model_cancer.feature_importances_
}).sort_values('importance', ascending=False)

# Визуализация топ-15 признаков
plt.figure(figsize=(14, 10))
top_n = 15
top_features = feature_importance_cancer.head(top_n)

colors_importance = plt.cm.RdYlGn(np.linspace(0.3, 0.9, top_n))
bars = plt.barh(range(top_n), top_features['importance'], color=colors_importance)
plt.yticks(range(top_n), top_features['feature'])
plt.xlabel('Важность признака', fontsize=12, fontweight='bold')
plt.title(f'Топ {top_n} наиболее важных признаков (Оптимизированный Random Forest)', 
          fontsize=14, fontweight='bold')
plt.gca().invert_yaxis()
plt.grid(True, alpha=0.3, axis='x')

# Добавление значений на график
for i, (bar, val) in enumerate(zip(bars, top_features['importance'])):
    plt.text(val + 0.001, bar.get_y() + bar.get_height()/2, 
             f'{val:.4f}', ha='left', va='center', fontsize=9)

plt.tight_layout()
plt.show()

print(f"\nТоп {top_n} наиболее важных признаков:")
for idx, row in feature_importance_cancer.head(top_n).iterrows():
    print(f"{row['feature']}: {row['importance']:.4f}")

# ============================================================================
# 5. ВЫВОДЫ И РЕКОМЕНДАЦИИ
# ============================================================================
print("\n" + "="*80)
print("5. ВЫВОДЫ И РЕКОМЕНДАЦИИ")
print("="*80)

best_model_name = cancer_results_df.iloc[0]['Model']
best_accuracy = cancer_results_df.iloc[0]['Test Accuracy']
best_cv = cancer_results_df.iloc[0]['CV Mean']

print(f"""
ОСНОВНЫЕ ВЫВОДЫ:

1. ПРОИЗВОДИТЕЛЬНОСТЬ МОДЕЛЕЙ:
   - Лучшая модель: {best_model_name}
   - Точность на тесте: {best_accuracy:.4f} ({best_accuracy*100:.2f}%)
   - CV точность: {best_cv:.4f} (±{cancer_results_df.iloc[0]['CV Std']:.4f})

2. СРАВНЕНИЕ ПОДХОДОВ:
   - Ансамблевые методы значительно превосходят одиночное дерево решений
   - Bagging методы (RF, Extra Trees) показывают стабильно высокие результаты
   - Boosting методы требуют тщательной настройки гиперпараметров

3. КЛЮЧЕВЫЕ ПРИЗНАКИ:
   - Наиболее важный признак: {feature_importance_cancer.iloc[0]['feature']}
   - Важность: {feature_importance_cancer.iloc[0]['importance']:.4f}
   - Топ-3 признака объясняют {feature_importance_cancer.head(3)['importance'].sum():.2%} дисперсии

4. РЕКОМЕНДАЦИИ:
   - Random Forest - оптимальный выбор для данной задачи (баланс точности и скорости)
   - Оптимизация гиперпараметров дала улучшение: {test_accuracy_optimized - cancer_results_df[cancer_results_df['Model'] == 'Random Forest']['Test Accuracy'].values[0]:.4f}
   - Для production рекомендуется использовать оптимизированную модель Random Forest
   - При необходимости максимальной точности можно рассмотреть стекинг ансамблей

5. КАЧЕСТВО КЛАССИФИКАЦИИ:
   - Модель достигла высокой точности на тестовой выборке
   - Отсутствие значительного переобучения (CV ~ Test Accuracy)
   - Модель готова к использованию в клинической практике
""")

print("="*80)
print("АНАЛИЗ ЗАВЕРШЕН")
print("="*80)
```
